Objective : Servo Prediction using Linear Regression

Data Source: Git Hub (https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Servo%20Mechanism.csv)

```python
#Step 1: Import Library
import pandas as pd
import numpy as np
```

```python
#Step 2: Import Data
servo = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Servo%20Mechanism.csv')
```

```python
#Step 3: Describe Data
servo.head()
```

```python
servo.info()
```

```python
servo.describe()
```

```python
servo.columns
```

```python
#Step 4: Encoding Categorial Features
servo.replace({'Motor':{'A':0,'B':1, 'C':2, 'D':3, 'E':4}},inplace=True)
servo.replace({'Screw':{'A':0,'B':1, 'C':2, 'D':3, 'E':4}},inplace=True)
```

```python
#Step 5: Define Y & X
y = servo['Class']
x = servo.drop('Class', axis=1)
```

```python
#Step 6: Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=2529)
```

```python
#Step 7: Model Train
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Step 8: Model Prediction
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#Step 9: Model Evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Step 10: Get Vizualization
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
